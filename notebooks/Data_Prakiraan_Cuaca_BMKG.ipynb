```python
from google.colab import userdata, auth
import os, requests, re, json, datetime
import matplotlib.pyplot as plt
import pandas as pd
import gspread
from gspread_dataframe import set_with_dataframe
from langchain_community.llms import Replicate
from IPython.display import display, Markdown
from google.auth import default
from collections import defaultdict

# === Konfigurasi Replicate API Token ===
api_token = userdata.get("api_token")

os.environ["REPLICATE_API_TOKEN"] = api_token

# === Pilih wilayah menggunakan API wilayah.id ===
province_code = "32"  # Jawa Barat
city_name_input = "Karawang"
url_city = f"https://wilayah.id/api/regencies/{province_code}.json"
resp_city = requests.get(url_city)

cities = resp_city.json().get("data", [])
kode_wilayah = next((c["code"] for c in cities if city_name_input.lower() in c["name"].lower()), None)

if not kode_wilayah:
    raise ValueError(f"Kota '{city_name_input}' tidak ditemukan di API wilayah.id.")

print("🌍===================================================")
print(f"📍 Prakiraan Cuaca Wilayah: {city_name_input} (Kode: {kode_wilayah})")
print("===================================================\n")

# === Ambil data BMKG ===
url_bmkg = f"https://api.bmkg.go.id/publik/prakiraan-cuaca?adm4={kode_wilayah}"
response = requests.get(url_bmkg, headers={"User-Agent": "Mozilla/5.0"})
data_text = response.text.strip()

today = datetime.date.today()

def build_dummy():
    """Dummy cuaca otomatis sesuai 3 hari ke depan"""
    dummy = []
    for d in range(3):
        base_date = (today + datetime.timedelta(days=d)).strftime("%Y-%m-%d")
        dummy += [
            {"local_datetime":f"{base_date} 00:00:00","t":"24","hu":"85","ws":10,"wd":"Timur","weather_desc":"Berawan","tcc":70},
            {"local_datetime":f"{base_date} 06:00:00","t":"26","hu":"80","ws":14,"wd":"Tenggara","weather_desc":"Cerah","tcc":20},
            {"local_datetime":f"{base_date} 12:00:00","t":"30","hu":"65","ws":22,"wd":"Barat Daya","weather_desc":"Hujan","tcc":100},
            {"local_datetime":f"{base_date} 18:00:00","t":"27","hu":"75","ws":15,"wd":"Barat Laut","weather_desc":"Berawan","tcc":60},
        ]
    return {"data":[{"cuaca": dummy}]}

if data_text.startswith("<!doctype html>") or "Akses diblokir" in data_text:
    print("⚠️ BMKG API diblokir. Menggunakan dummy data simulasi.\n")
    cuaca_data = build_dummy()
else:
    try:
        cuaca_data = response.json()
        if not cuaca_data.get("data"):  # jika kosong fallback ke dummy
            cuaca_data = build_dummy()
    except Exception:
        json_match = re.search(r"\{.*\}", data_text, re.DOTALL)
        cuaca_data = json.loads(json_match.group()) if json_match else build_dummy()

# === Flatten data ===
prakiraan_all = []
cuaca_list = cuaca_data.get("data", [])
if cuaca_list:
    for item in cuaca_list[0].get("cuaca", []):
        prakiraan_all.append(item if isinstance(item, dict) else item[0])

# === Ambil 3 hari ke depan ===
unique_dates = sorted({d["local_datetime"][:10] for d in prakiraan_all})
next_3days = [d for d in unique_dates if datetime.datetime.strptime(d, "%Y-%m-%d").date() >= today][:3]
filtered = [d for d in prakiraan_all if d["local_datetime"][:10] in next_3days]

# ✅ Kelompokkan data per tanggal
grouped = defaultdict(list)
for item in filtered:
    tanggal = item["local_datetime"][:10]
    grouped[tanggal].append(item)

days = [grouped[t] for t in sorted(grouped.keys())[:3]]

if not days:  # fallback terakhir
    days = [build_dummy()["data"][0]["cuaca"][i:i+4] for i in range(0, 12, 4)]

# === Fungsi format tanggal ===
def format_tanggal(tanggal_str):
    dt = datetime.datetime.strptime(tanggal_str, "%Y-%m-%d %H:%M:%S")
    hari_indo = {
        "Monday": "Senin","Tuesday": "Selasa","Wednesday": "Rabu",
        "Thursday": "Kamis","Friday": "Jumat","Saturday": "Sabtu","Sunday": "Minggu"
    }[dt.strftime("%A")]
    return f"{hari_indo}, {dt.strftime('%d %B %Y %H:%M')}"

# === Klasifikasi ===
def classify_day(day):
    weather_list = [d["weather_desc"] for d in day]
    max_wind = max(d["ws"] for d in day)
    rain_count = sum("Hujan" in w for w in weather_list)
    if rain_count > 2: kategori = "Hujan"
    elif weather_list.count("Cerah") > 2 or weather_list.count("Cerah Berawan") > 2: kategori = "Cerah"
    else: kategori = "Berawan"
    ekstrem = []
    if max_wind > 20: ekstrem.append(f"🌬️ Angin kencang {max_wind} km/jam")
    if rain_count/len(day) > 0.5: ekstrem.append("☔ Hujan dominan")
    return kategori, ekstrem

# === Setup Granite ===
llm = Replicate(
    model="ibm-granite/granite-3.3-8b-instruct",
    model_kwargs={"temperature":0.3, "max_tokens":500}
)

# === Loop per Hari ===
all_summary, all_days_df = "", []

for idx, day in enumerate(days, start=1):
    display(Markdown(f"## 📅 Hari ke-{idx} ({day[0]['local_datetime'][:10]})"))
    display(Markdown("---"))

    records, text_data = [], ""
    for item in day:
        records.append({
            "Hari": idx,
            "Waktu": format_tanggal(item["local_datetime"]),
            "🌡️ Suhu (°C)": int(item["t"]),
            "💧 Kelembapan (%)": int(item["hu"]),
            "🌬️ Angin (km/jam)": int(item["ws"]),
            "☁️ Cuaca": item["weather_desc"],
            "Awan (%)": int(item["tcc"])
        })
        text_data += f"- {format_tanggal(item['local_datetime'])} | 🌡️ {item['t']}°C | 💧 {item['hu']}% | {item['weather_desc']} | 🌬️ {item['ws']} km/jam ({item['wd']}) | ☁️ {item['tcc']}%\n"

    df = pd.DataFrame(records)
    all_days_df.append((f"Hari-{idx}", df))

    # 🎨 Tabel styled
    styled_df = (
        df.style.set_caption(f"Tabel Cuaca Hari ke-{idx}")
        .format({
            "🌡️ Suhu (°C)": "{:.0f}",
            "💧 Kelembapan (%)": "{:.0f}",
            "🌬️ Angin (km/jam)": "{:.0f}",
            "Awan (%)": "{:.0f}"
        })
        .background_gradient(subset=["🌡️ Suhu (°C)"], cmap="Oranges")
        .background_gradient(subset=["💧 Kelembapan (%)"], cmap="Blues")
        .background_gradient(subset=["🌬️ Angin (km/jam)"], cmap="Greens")
    )
    display(styled_df)

    # 📈 Grafik tren (hanya 1 chart per hari)
    plt.figure(figsize=(10,5))
    plt.plot(df["Waktu"], df["🌡️ Suhu (°C)"], marker="o", color="orange", label="Suhu (°C)")
    plt.plot(df["Waktu"], df["💧 Kelembapan (%)"], marker="s", color="blue", label="Kelembapan (%)")
    plt.plot(df["Waktu"], df["🌬️ Angin (km/jam)"], marker="^", color="green", label="Angin (km/jam)")
    plt.title(f"Tren Cuaca Hari ke-{idx}")
    plt.xticks(rotation=30)
    plt.xlabel("Waktu")
    plt.ylabel("Nilai")
    plt.legend()
    plt.grid(True)
    plt.show()

    # 📝 Ringkasan AI
    kategori, ekstrem = classify_day(day)
    accuracy = 90 if kategori=="Cerah" else 80 if kategori=="Berawan" else 70
    ekstrem_text = ", ".join(ekstrem) if ekstrem else "Normal ✅"

    prompt = f"""
Ringkas prakiraan cuaca BMKG untuk wilayah {city_name_input} hari ke-{idx}.
Gunakan bahasa santai dengan emoji agar menarik dibaca.
Sertakan kategori cuaca: {kategori}, indikator ekstrem: {ekstrem_text}, perkiraan akurasi: {accuracy}%.
Data:
{text_data}
"""
    summary = llm.invoke(prompt)

    display(Markdown(f"""
### 📝 Ringkasan AI (Hari {idx})
- **Kategori:** {kategori}
- **Ekstrem:** {ekstrem_text}
- **Akurasi:** {accuracy}%

{summary}
    """))

    all_summary += f"Hari {idx}: {summary}\n"

# === Export ke Google Spreadsheet (Per Hari) ===
if all_days_df:
    auth.authenticate_user()
    creds, _ = default()
    gc = gspread.authorize(creds)

    try:
        sh = gc.open("Prakiraan Cuaca 3 Hari")
    except gspread.SpreadsheetNotFound:
        sh = gc.create("Prakiraan Cuaca 3 Hari")

    for sheet_title, df_day in all_days_df:
        try:
            ws_day = sh.worksheet(sheet_title)
        except:
            ws_day = sh.add_worksheet(title=sheet_title, rows=100, cols=20)
        ws_day.clear()
        set_with_dataframe(ws_day, df_day)

    try:
        ws_summary = sh.worksheet("Ringkasan AI")
    except:
        ws_summary = sh.add_worksheet(title="Ringkasan AI", rows=100, cols=10)

    ws_summary.clear()
    ws_summary.update(values=[["Hari", "Ringkasan"]], range_name="A1")
    summary_rows = [line for line in all_summary.split("\n") if line.strip()]
    ws_summary.update(
        values=[[f"Hari {i+1}", s] for i, s in enumerate(summary_rows)],
        range_name="A2"
    )

    print(f"✅ Semua data sudah masuk ke Google Spreadsheet: {sh.url}")

# === Kesimpulan umum
display(Markdown("## 🌈 Kesimpulan Umum 3 Hari ke Depan"))
display(Markdown("---"))

final_prompt = f"""
Berdasarkan prakiraan cuaca BMKG 3 hari ke depan untuk {city_name_input}:

{all_summary}

Buat kesimpulan umum dengan gaya bahasa ramah + emoji.
Berikan juga saran singkat.
"""
final_output = llm.invoke(final_prompt)
display(Markdown(f"### 🌟 Ringkasan Umum\n\n{final_output}"))
```
